# Importing Library


```python
import pandas as pd
```

# Importing Data

```python
df1=pd.read_excel(r"C:\Users\alika\Downloads\Customer Call List.xlsx")
df1
```

# Removing N/A values

```python
df1.fillna('')
```

# Dropping unwanted columns

```python
df1=df1.drop(columns='Not_Useful_Column')
```

# Dropping duplicate rows

```python
df1=df1.drop_duplicates()
```

# Using strip function to remove unwanted characters

```python
df1["Last_Name"]=df1["Last_Name"].str.strip("/.-_")
df1
```

# Converting "Phone_Number" column to string type

```python
df1["Phone_Number"]=df1["Phone_Number"].apply(lambda x: str(x))
```

# Removing non alphabetical and non numeric values from "Phone_Number"

```python
df1["Phone_Number"]=df1["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
df1
```

# Formatting the "Phone_Number" column

```python
df1["Phone_Number"]=df1["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df1
```

# Removing blank cells from the column

```python
df1["Phone_Number"]=df1["Phone_Number"].str.replace('--','')
df1["Phone_Number"]=df1["Phone_Number"].str.replace('Na','')
df1["Phone_Number"]=df1["Phone_Number"].str.replace('nan','')
df1
```

# Splitting the "Address" column

```python
df1[["House_Number","Add"]]=df1["Address"].str.split(' ',1,expand=True)
df1=df1.drop(columns="Address")
df1
```

# Splitting the "Add" column

```python
df1[["Street","City","Postal_Code"]]=df1["Add"].str.split(',',2,expand=True)
df1=df1.drop(columns="Add")
df1
```

# Removing unwanted characters from "House_Number" column

```python
df1["House_Number"]=df1["House_Number"].str.strip('thN/a')
df1
```

# Reordering columns

```python
df1=df1.reindex(columns=["CustomerID","First_Name","Last_Name","Phone_Number","House_Number","Street","City","Postal_Code","Paying Customer","Do_Not_Contact"])
df1
```

# Formatting the "City" column

```python
df1["City"]=df1["City"].apply(lambda x: str(x))
df1["City"]=df1["City"].str.replace('None','-')
df1
```

# Formatting the "Postal_Code" column

```python
df1["Postal_Code"]=df1["Postal_Code"].apply(lambda x: str(x))
df1["Postal_Code"]=df1["Postal_Code"].str.replace('None','-')
df1
```

# Formatting the "Paying_Customer" column

```python
df1["Paying Customer"]=df1["Paying Customer"].str.replace('No','N')
df1
```

```python
df1["Paying Customer"]=df1["Paying Customer"].str.replace('Yes','Y')
df1
```

```python
df1["Paying Customer"]=df1["Paying Customer"].str.replace('N/a','-')
df1
```

# Formatting "Do_Not_Contact" column

```python
df1["Do_Not_Contact"]=df1["Do_Not_Contact"].str.replace('Yes','Y')
df1
```

```python
df1["Do_Not_Contact"]=df1["Do_Not_Contact"].str.replace('No','N')
df1
```

```python
df1["Do_Not_Contact"]=df1["Do_Not_Contact"].apply(lambda x: str(x))
```

```python
df1["Do_Not_Contact"]=df1["Do_Not_Contact"].str.replace('nan','-')
df1
```

# Formatting "Street" column

```python
df1["Street"]=df1["Street"].apply(lambda x: str(x))
```

```python
df1["Street"]=df1["Street"].str.replace('None','-')
df1
```

# Combining two columns

```python
value=df1["First_Name"]+' '+df1["Last_Name"]
value
```

# Inserting new column "Full_Name"

```python
df1.insert(loc=1,column='Full_Name',value=value)
df1
```

# Dropping non useful columns

```python
df1.drop(columns=["First_Name","Last_Name"],inplace=True)
```

# Converting the new column into string type

```python
df1["Full_Name"]=df1["Full_Name"].apply(lambda x: str(x))
```

```python
df1["Full_Name"]=df1["Full_Name"].str.replace('nan','-')
```

```python
df1
```

# Formatting "Do_Not_Contact" column

```python
for x in df1.index:
    if df1.loc[x,'Do_Not_Contact']=='Y':
        df1.drop(x,inplace=True)
df1
```

```python
df1["Do_Not_Contact"]=df1["Do_Not_Contact"].str.replace('-','N')
df1
```

# Formatting "Phone_Number" column

```python
for x in df1.index:
    if df1.loc[x,'Phone_Number']=='':
        df1.drop(x,inplace=True)
```

```python
df1.drop(columns=df1.columns[6],inplace=True)
```
